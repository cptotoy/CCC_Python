```python
# Messy data from spreadsheet
raw_data = [' JOHN SMITH ', ' alice JOHNSON', 'BOB jones']
```

```python
# Clean it up!
clean_data = []
for name in raw_data:
    clean = name.strip().title()
    clean_data.append(clean)
print(clean_data)
```

```python
# Alternative — handle multiple internal spaces:
for name in raw_data:
    clean = " ".join(name.split()).title()
    clean_data.append(clean)
print(clean_data)
```
