```python
email1 = 'user@company.com'
email2 = 'invalid.email'
email3 = ' USER@EMAIL.COM '
```

```python
email = email3.strip().lower()

if '@' in email and '.' in email:
    print('Valid')
    at = email.index('@')
    user = email[:at]
    domain = email[at+1:]
    print(f'User: {user}')
    print(f'Domain: {domain}')
else:
    print('Invalid')
```
