```python
text = 'I am learning Python and I love it!'
```

```python
#1. Total characters (including spaces) → 
Len_txt= len(text)
#2. Total characters (no spaces) → l
total_chars=len(text.replace(' ', ''))
#3. Number of words → 
word_count=len(text.split())
#4. Average word length → 
Average = total_chars / word_count
#5. Uppercase version → 
Upper= text.upper()
#6. Lowercase version → 
Lower=text.lower()
```

```python
print(Len_txt)
print(total_chars)
print(word_count)
print(Average)
print(Upper)
print(Lower)
```
