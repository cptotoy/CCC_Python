Nested Dictionaries

```python
# Flat dict – one level
student = {'name': 'Alice', 'age': 20, 'grade': 'A'}

# Nested dict – multiple levels
school = {
    'Alice': {'age': 20, 'grade': 'A', 'gpa': 3.9},
    'Bob':   {'age': 21, 'grade': 'B', 'gpa': 3.2},
    'Carol': {'age': 19, 'grade': 'A', 'gpa': 3.7},
}

# Access nested values – chain the keys
print(school['Alice']['gpa'])          # 3.9
print(school['Bob']['grade'])          # B

# PCEP trap – wrong nesting order
# school['gpa']['Alice'] -> KeyError! Keys are names, not fields
```

Accessing nested values safely

```python
school = {
    'Alice': {'age': 20, 'grade': 'A', 'gpa': 3.9},
    'Bob':   {'age': 21, 'grade': 'B', 'gpa': 3.2},
}

# Direct access – crashes if either key missing
print(school['Alice']['gpa'])                          # 3.9

# Safe access with .get() – first level
alice = school.get('Alice', {})
print(alice.get('gpa', 'N/A'))                         # 3.9

# Safe access chained – handles BOTH levels missing
print(school.get('Dave', {}).get('gpa', 'N/A'))       # N/A – Dave not in school

# Modifying a nested value
school['Alice']['gpa'] = 4.0                           # update Alice's gpa

# Adding a new nested entry
school['Dave'] = {'age': 22, 'grade': 'C', 'gpa': 2.8}
```

Iterating over nested dictionaries

```python
school = {
    'Alice': {'age': 20, 'grade': 'A', 'gpa': 3.9},
    'Bob':   {'age': 21, 'grade': 'B', 'gpa': 3.2},
    'Carol': {'age': 19, 'grade': 'A', 'gpa': 3.7},
}

# Iterate outer dict – name and inner dict
for name, info in school.items():
    print(f'{name}: grade={info["grade"]}, gpa={info["gpa"]}')

# Iterate inner dict for each student
for name, info in school.items():
    print(f'*** {name} ***')
    for field, value in info.items():
        print(f'   {field}: {value}')

# Collect all GPAs into a list
all_gpas = [info['gpa'] for name, info in school.items()]
print(f'Average GPA: {sum(all_gpas)/len(all_gpas):.1f}')
```

Dictionarioes with list values - mized structures

```python
# Dict where each value is a LIST
courses = {
    'Alice': ['Math', 'CS', 'Physics'],
    'Bob':   ['Math', 'English'],
    'Carol': ['CS', 'Art', 'Music', 'History'],
}

# Access: outer key then list index
print(courses['Alice'][0])         # Math  (first course)
print(courses['Carol'][-1])        # History (last course)

# Append to a student's course list
courses['Bob'].append('CS')        # Bob now has 3 courses

# Iterate: student and their course list
for student, course_list in courses.items():
    print(f'{student} takes {len(course_list)} courses')

# Find students enrolled in 'CS'
cs_students = [s for s, c in courses.items() if 'CS' in c]
print(cs_students)  # ['Alice', 'Carol', 'Bob']
```

TUPLES

```python
# Creating tuples - use parentheses
point = (3, 5)                  # tuple with 2 items

# PCEP TRAP - single-element tuple needs a trailing comma
not_a_tuple = (5)               # just an int - NOT a tuple!
one_item = (5,)                 # comma makes it a tuple
empty = ()                      # empty tuple is fine
# Parentheses are OPTIONAL - comma is what creates a tuple
colors = 'red', 'green', 'blue' # still a tuple - print(type(colors))

# Indexing and slicing - same as lists
print(point[0])                 # 3
print(colors[1:])               # ('green', 'blue')

# Traversing - same as lists
for c in colors: print(c)       # red, green, blue
```

Tuple Unpacking & Multiple Return Values

```python
# Unpacking - assign tuple items to variables in one step
point = (3, 5)
x, y = point  # x=3, y=5
print(point)
# PCEP TRAP - count must match exactly

# a, b, c = point   -> ValueError: not enough values to unpack
# Swap variables in one line - Python idiom

a, b = 10, 20
a, b = b, a  # a=20, b=10  (no temp variable!)
# Functions can return multiple values - they come back as a tuple


def min_max(nums):
    return min(nums), max(nums)


low, high = min_max([4, 9, 2, 7])
print(low, high)  # 2 9

# Adding a new nested entry
school["Dave"] = {"age": 22, "grade": "C", "gpa": 2.8}
```

TUPLES vs LISTS

```python
# Lists are MUTABLE - you can change them
my_list = [1, 2, 3]
my_list[0] = 99                 # works! -> [99, 2, 3]
my_list.append(4)               # works! -> [99, 2, 3, 4]
# Tuples are IMMUTABLE - you CANNOT change them

my_tup = (1, 2, 3)
# my_tup[0] = 99        -> TypeError: tuple does not support assignment
# my_tup.append(4)      -> AttributeError: tuple has no append method

# PCEP EXAM - but the contents of a mutable item INSIDE a tuple CAN change
mixed = ([1, 2], [3, 4])        # tuple containing two lists
mixed[0].append(99)             # works! tuple unchanged, list changed
print(mixed)                    # ([1, 2, 99], [3, 4])
# mixed[0] = [9]        -> TypeError - cannot REASSIGN tuple slots

# Why tuples exist: speed, safety, can be dict keys
# Use a tuple when the collection should NEVER change
# Use a list when you need to add, remove, or modify items
```

Tuples as dict Keys - Unique Identifiers

```python
# Lists CANNOT be dict keys (they are mutable)
# Tuples CAN be dict keys (they are immutable)
# Use tuples to key on a combination of values
locations = {
    (40.7, -74.0):  'New York',
    (41.8, -87.6):  'Chicago',
    
    (34.0, -118.2): 'Los Angeles',
}
print(locations[(41.8, -87.6)])       # Chicago

# PCEP TRAP — what happens if you try a list as a key?
# bad = { [1, 2]: 'point' }    -> TypeError: unhashable type: list

# Use case: grid coordinates, multi-field IDs, frequency counts
scores = {}
scores[('Alice', 'Math')] = 95         # student-subject pair as key

scores[('Alice', 'CS')]   = 88
scores[('Bob', 'Math')]   = 72
print(scores[('Alice', 'Math')])       # 95
```

JSON

```python
import json

# json.loads() - parse a JSON STRING into a Python dict
json_text = '{"name":"Alice","age":20,"scores":[95,87,92]}'
data = json.loads(json_text)
print(data['name'])          # Alice
print(data['scores'][1])     # 87

# json.dumps() - convert a Python dict to a JSON STRING
student = {'name': 'Bob', 'gpa': 3.5, 'active': True}
json_string = json.dumps(student)
print(json_string)  # {"name": "Bob", "gpa": 3.5, "active": true}

# Pretty-print with indent
print(json.dumps(student, indent=2))

# json.load() / json.dump() - read/write FILES
with open('data.json', 'r') as f:
    data = json.load(f)      # file -> dict
with open('output.json', 'w') as f:
    json.dump(data, f, indent=2)  # dict -> file
```
