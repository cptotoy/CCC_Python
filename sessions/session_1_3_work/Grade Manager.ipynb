Grade Manager

```python
grades = []
grades.append(95)
grades.append(87)
grades.append(92)
grades.append(78)
grades.append(88)
print(grades)
```

```python
#Number of grades
Len_grades =len(grades)
#Highest grade
Max_grades = max(grades)
#Lowest grade
Min_grades = min(grades)
#Average grade
Avg_grade = sum(grades)//len(grades)
#print
print(f'Number of Grades: {Len_grades}\nHighest grade: { Max_grades}\nLowest Grade: {Min_grades}\nAverage grade: {Avg_grade}')
```

```python
grades.sort()
print(grades)
```
