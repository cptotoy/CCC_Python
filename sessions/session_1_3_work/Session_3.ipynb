```python
# Like strings, lists use zero-based positions. The first item is always index 0

students =['Alice', 'Bob', 'Carol', 'David', 'Eve']
print(students[0])  #Alice -> First item 
print(students[2])  #Carol -> Third item
print(students[-1]) #Alice -> last item item
print(students[-2]) #Alice -> before last item
```

```python
scores  = [95,87,92,78,88,91]
count   = len(scores)
total   = sum(scores)
average = total/count
highest = max(scores)
lowest  = min(scores)

print(f'Scores:{count}  |   Avg:{average:.1f}   |   High: {highest}  |   low {lowest}')
```
