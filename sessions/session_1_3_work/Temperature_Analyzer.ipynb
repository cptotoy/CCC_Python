Temperature Analizer

```python
temperatures = [72, 68, 75, 70, 73, 69, 71]
```

```python
avg_temp = sum(temperatures)//len(temperatures)
#Highest temperature
Max_temp = max(temperatures)
#Lowest temperatures
Min_temp = min(temperatures)
#Temperature range
Range_temp = Max_temp-Min_temp
#Days above average
Count_temp = len([t for t in temperatures if t > avg_temp])
#Days below 70
days_below_70 = len([t for t in temperatures if t < 70])
```

```python
print(avg_temp)
#Highest temperature
print(Max_temp)
#Lowest temperatures
print(Min_temp)
#Temperature range
print(Range_temp)
#Days above average
print(Count_temp)
#Days below 70
print(days_below_70)
```
