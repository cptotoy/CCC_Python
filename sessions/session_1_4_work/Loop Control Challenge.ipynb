```python
nums =[4,7,2,9,1,8,3,6,5]

for n in nums:
    if n==9:
        continue
    print(n, end=' ')
```

```python
for n in nums:
    if n==8:
        break
    print(n, end=' ')
```

```python
count = 0
for n in nums:
    if n > 5:
        count += 1

print(f'Count of numbers: {count}')
```
