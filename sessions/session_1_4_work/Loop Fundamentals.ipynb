```python
i = 1
square = 0

while square <= 50:
    square = i ** 2
    if square > 50:
        break
    print(f"{i} squared is {square}")
    i += 1
```

```python
total = 0
i = 1

while i <= 100:
    total += i
    i += 1

print("The sum is:", total)
```
