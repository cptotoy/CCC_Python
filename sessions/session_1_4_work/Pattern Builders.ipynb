```python
# Pattern 1 — multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(i * j, end=' ')
    print()
```

```python
# Pattern 2 — grade classifier
scores = [88, 72, 95, 61, 83]
for s in scores:
    if s >= 90: grade = 'A'
    elif s >= 80: grade = 'B'
    elif s >= 70: grade = 'C'
    else: grade = 'F'
    print(f'{s} → {grade}')
```

```python
# Pattern 3 — accumulate until target
data = [5, 3, 8, 2, 9, 4, 7]
running = 0
for val in data:
    running += val
    print(f'Running total: {running}')
    if running >= 20:
        print('Target reached!')
        continue
    break
```
