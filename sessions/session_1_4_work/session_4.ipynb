```python
daily_sales = [120, 85, 200, 95, 175, 60, 230]

# Pattern 1 - Accumulator (buid a running total)
revenue = 0 
for sale in daily_sales:
    revenue += sale
    print(f'Weekly revenue: ${revenue}')
```

```python
# Pattern 2 - Counter (count days above target)
good_days = 0 
for sale in daily_sales:
    if sale>=150:
        good_days +=1
print(f'Days above target: {good_days}')
```

```python
# Pattern 3 - Collector (build filtered list)
slow_days = []
for sale in daily_sales:
    if sale < 100:
        slow_days.append(sale)
print(f'Slow days: {slow_days}')
```

LOOP COntrol: break

```python
scores =[72, 85, 91, 78, 95]
```

```python
# find first score >=90
for score in scores: 
    if score>=90:
        print(f'Found:{score}')
        break

# Found : 91
# Stops after finding first match 
```

```python
# Search trough a list
names = ['Alice', 'Christian', 'Carol']
target = 'Christian'

for name in names:
    if name == target:
        print('Found target!')
        break
else:
    print('Not found')
```

```python
for i in range(3):
    if i == 2:
        break
else:
    print('done')
print('end')

# Output? -> end (else skipped - break fired)
```

```python
i = 0
while i < 4:
    i += 1
    if i ==2:
        continue
    print(i, end=' ')
    
# Output? -> 1 3 4 ( 2 skipped by continue)
```
