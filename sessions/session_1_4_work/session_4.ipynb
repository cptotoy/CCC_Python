Loops

```python
def process(reading):
    print(reading)

readings = [23.1, 24.5, 22.8, 21.9, 24.1, 12.3]
process(readings[0])
process(readings[1])
process(readings[2])
# ... 97 more lines for 100 readings!
```

```python
def process(jorge):
    print(jorge)

for jorge in readings: 
    process(jorge)
```
