```python
age = 20 
has_id = False

# and - Both must be true
if age >= 18 and has_id:
        print('entry allowed')

# or - AT LEAST ONE must be Trie
if age < 18 or not has_id:
        print('entry denied')

# not - flips TRUE to FALSE, FALSE to TRUE
if not has_id: 
    print('No ID- denied')
```

```python
#1. Grade classifier - chained elif
def classify_grade(score):
    if score >= 90:     return 'A - Excellent!'
    elif score >= 80:   return 'B - Excellent!'
    elif score >= 70:   return 'C - Excellent!'
    elif score >= 60:   return 'D - Excellent!'
    else:               return 'F - Plase retake'
print(classify_grade(40))
```

```python
age = 20
has_ticket = True
is_vip = False

if age >=18: 
    print('Age check: OK')
    if has_ticket:
        print('has ticket: OK')
        if is_vip:
            print('VIP seat!')
        else: 
            print('General seat')
    else:
        print('No ticket- denied')
else:
    print('Under 18 - denied')

#-   Each level adds 4 more spaces
#-   Inner if only runs if outer s TRUE
#-   Can go as deep as needed (but 2-3 levels is ussually enough!)
#-   A good tip is to add an ELSE statement below the if, so we dont get lost
```

```python
# Precedence in practice - trace

# Without parentheses - what does python actually do?

x = 5
y = 3
z = 10
# Example 1: and before or
result = x>3 or y> 5 and z> 8
#Python read: x>3 or (y>5 and z>8)
# Step 1: y>5 and z>8 = False and True = False
# Step 2:x>3 or FALSE = TRUE or FALSE = TRUE
print(result)
```

```python
# Example 2: not before and
result2 = not x>10 and y>1
#Python reads: (not x>10) and y>1
#Step 1: not x>10 = not FALSE = TRUE
#Step 2: True and y>1 = TRUE and TRUE= TRUE
print(result2)
```

```python
#traditional if/false:

score = 85
if score >= 60:
    result = 'Pass'
else:
    result = 'Fail'
print(result)

# Ternary equivalent (same result, one line):
result = 'Pass' if score >60 else 'Fail'
print(result)

#Works in list comprehensions too:

scores = [85,42,91,55]
labels = ['Pass' if s>=60 else 'Fail' for s in scores]
print(labels)
```
