```python
# File: grade_matrix.py

grades = [[95, 87, 92, 88], [92, 88, 90, 85], [72, 68, 79, 75]]
students = ["Alice", "Bob", "Carol"]
subjects = ["Math", "English", "Science", "History"]

print("--- Data Analysis Results ---")

# 1. Average score per student (row analysis)
print("\n1. Average Score per Student:")
for i in range(len(students)):
    student_total = 0  # Accumulator for row values
    for j in range(len(subjects)):
        student_total += grades[i][j]
    student_avg = student_total / len(subjects)
    print(f"   {students[i]}: {student_avg:.2f}")

# 2. Average score per subject (column analysis)
print("\n2. Average Score per Subject:")
for j in range(len(subjects)):
    subject_total = 0  # Accumulator for column values
    for i in range(len(students)):
        subject_total += grades[i][j]
    subject_avg = subject_total / len(students)
    print(f"   {subjects[j]}: {subject_avg:.2f}")

# Initialize global accumulators for metrics 3, 4, and 5
grand_total = 0
total_grades_count = 0
highest_score = -1
highest_student = ""
highest_subject = ""
a_grades_count = 0

# Flattened nested loop to sweep the entire matrix
for i in range(len(students)):
    for j in range(len(subjects)):
        current_grade = grades[i][j]

        # Accumulate values for overall average
        grand_total += current_grade
        total_grades_count += 1

        # Conditional pattern to find the maximum grade
        if current_grade > highest_score:
            highest_score = current_grade
            highest_student = students[i]
            highest_subject = subjects[j]

        # Conditional pattern to filter and count 'A' grades
        if current_grade >= 90:
            a_grades_count += 1

# 3. Overall class average
overall_average = grand_total / total_grades_count
print(f"\n3. Overall Class Average: {overall_average:.2f}")

# 4. Highest score in entire matrix
print(
    f"4. Highest Score: {highest_score} (Earned by {highest_student} in {highest_subject})"
)

# 5. Count of A grades
print(f"5. Total Count of 'A' Grades (>= 90): {a_grades_count}")
```
