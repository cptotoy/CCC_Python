```python

# Print the top column headers
print("    ", end="")  # Initial spacing for the top-left corner
for col in range(1, 11):
    print(f"{col:4d}", end="")
print("\n" + "    " + "----" * 10)  # Visual separator line

# Nested loops to build the table rows
for row in range(1, 11):
    # Print the row label at the start of each row
    print(f"{row:2d} |", end="")
    
    # Inner loop calculates and prints each column value
    for col in range(1, 11):
        product = row * col
        print(f"{product:4d}", end="")
    
    # Move to the next line after completing a row
    print()

```
