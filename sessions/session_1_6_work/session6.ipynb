```python
grades = [
    [95, 87, 92],
    [88, 90, 85],
    [72, 68, 79],    
]
names = ['Alice','Bob','Carol']

for student_idx, student_grades in enumerate(grades):
    total = 0 
    for grade in student_grades:
        total += grade
    average = total/len(student_grades)
    status = 'Pass' if average>=70 else 'Fail'
    print(f'{names[student_idx]}: avg={average:1f} {status}')
```

```python
colors = ['red', 'blue', 'green']
sizes = ['small','medium', 'large', 'XL']
products=[]

for color in colors:
    for size in sizes:
        if size !='XL' or color == 'blue': 
            products.append(f'{size} {color}')

print(f'Products available: {len(products)}')
```

```python
matrix = [[1, 2], [3, 4], [5, 6]]
total = 0
for row in matrix:
    for val in row:
        if val % 2 == 0:
            total += val
print(total)
```

```python
data = [[3, -1, 4], [-2, 7, -5], [1, 6, -3]]
total = 0
for row in data:
    for val in row:
        if val > 0:  # (1) Correct insertion
            total += val
print(total)
```
