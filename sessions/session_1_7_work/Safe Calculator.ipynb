```python
while True:
    try:
        # 1. Ask for the first input as text so we can check for 'quit'
        first_input = input('Enter first number (or type \'quit\'): ')
        if first_input == 'quit':
            print("Goodbye!")
            break
        Number_1 = int(first_input)
        
        second_input = input('Enter second number (or type \'quit\'): ').strip().lower()
        if second_input == 'quit':
            print("Goodbye!")
            break
        Number_2 = int(second_input)
        
        
        print(f'Your numbers are {Number_1} and {Number_2}')
        
        # 2. Get and validate the operator inside the try block
        operator = input("Enter an operator (+, -, *, /) (or type 'quit'): ").strip().lower()
        
        if operator == 'quit':
            print("Goodbye!")
            break
        elif operator not in ['+', '-', '*', '/']:
            raise ValueError("UnknownOperator")
            
        # 3. Perform calculation using distinct if/elif blocks
        if operator == '+':
            result = Number_1 + Number_2
        elif operator == '-':
            result = Number_1 - Number_2
        elif operator == '*':
            result = Number_1 * Number_2
        elif operator == '/':
            result = Number_1 / Number_2  # Triggers ZeroDivisionError if Number_2 is 0

        # 5. Print the result formatted to exactly 2 decimal places
        print(f"Result: {result:.2f}")

    # 4. Handle ALL error cases safely without crashing
    except ValueError as e:
        if str(e) == "UnknownOperator":
            print(f"Error: '{operator}' is not a valid operator.")
        else:
            print('Error: Input must be a valid number, not text.')
            
    except ZeroDivisionError:
        print('Error: Cannot divide by zero!')
```
