```python
try: 
    #Code that might raise an error
    risky_operation()
except Errortype:
    # Runs ONLY if that error occurs
    handle_the_error()
else:
    # Runs ONLY if NO error occurred
    success_operation()
finally:
    # Always runs - error or not 
    cleanup_operation()
```

```python
# Example 1: Catching a specific error
try: 
    result = 10/0
except ZeroDivisionError:
    print('Cannot divide by zero')
print('Program continues...')
```

```python
# Example 2: Catching ValueError from bad input
try:
    age = int(input ('Enter your age: '))
    print( f'You are {age} years old')
except ValueError:
    print('Age must be a whole number!')
```

```python
# Example 3: Multiple except blocks
try:
    nums = [1,2,3]
    result = nums [10]/0
except IndexError:
    print('List index out of range!')
except ZeroDivisionError:
    print('Cannot divide by zero!')
print('Program continues...')
```

```python
def safe_divide(a,b):
    try:
        result = a / b
    except ZeroDivisionError:
        print('Error: Cannot divide by zero!')
        result = None
    else:
        print(f'Success: {a} / {b} = {result:.2f}')
    finally:
        print('Division attempt complete.')
    return result
safe_divide(10,2)
#safe_divide(10,0)
```

```python
# Strategy 1: Multiple except blocks (preferred – most specific)
try:
    value = int(input('Enter index: '))
    nums = [10, 20, 30]
    print(nums[value])
except ValueError:
    print('Please enter a whole number.')
except IndexError:
    print('Index out of range – list has 3 items.')

```

```python
# Strategy 2: Catch multiple in one except
try:
    result = int(input('Number: ')) / int(input('Divisor: '))
except (ValueError, ZeroDivisionError) as e:
    print(f'Input error: {e}')  # 'as e' captures the message
print('Division attempt complete.')
```

```python
# Strategy 3: Bare except (use sparingly – catches everything)
try:
    risky_code()
except Exception as e:          # Catches all standard exceptions
    print(f'Unexpected error: {e}')
```
