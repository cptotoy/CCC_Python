```python
# // Floor division - rounds Down to nearest integer
print (7//2) 
print (-7//2)
print(7.0//2)
```

```python
# % Modilo - remainder after division
print (10 % 3)
print (7 % 7)
print (-7 % 3)
```

```python
# ** Exponentiation - higher precedence than unary minus!
print (2** 10)
print (-2 ** 2)
print((-2)** 2)

# PCEP trap: -2 ** 2 -> -(2**2) = -4 NOT (-2)**2 = 4 
```

Bitwise operators work on indificual bits of integers

```python
# a = 6 -> 0110 in binary
# b = 3 -> 0011 in binary 
a= 6 ; b = 3

print(a & b)    # and: 0110 & 0011 = 0010 = 2
print (a | b)   # ORL 0110 | 0011 = 0111 = 7
print (a ^ b)   # XOR: 0110 ^ 0011 = 0101 = 5
print (~a)      # NOT: ~0110 = -(a+1) = -7
print (a << 1)  # Left shift: 0110 -> 1100 = 12
print (a >> 1)  # Right Shift: 0110 -> 0011 = 3
```

```python
# is / is not - Identity (same object in memory)
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print(a == b)       # True  - same VALUES
print(a is b)       # False - different objects in memory
print(a is c)       # True  - c points to the SAME object as a

# None check - ALWAYS use 'is', never '=='
result = None
if result is None:       # Correct
    print('No result')

# in / not in - Membership (does item exist in sequence?)
fruits = ['apple', 'banana', 'cherry']
print('apple' in fruits)       # True
print('mango' not in fruits)   # True

text = 'Hello Python'
print('Python' in text)        # True  - works on strings too
print('python' in text)        # False - case-sensitive!
```

```python
#Explicitly converting between types. A major source of runtime errors — and PCEP questions.

# int() — converts to integer
print(int('42'))       # 42        — string of digits -> int
print(int(3.9))        # 3         — float -> int (TRUNCATES, does not round)
print(int(True))       # 1         — bool -> int
# int('3.5')                       # ValueError! — string with decimal point fails

# float() — converts to float
print(float('3.14'))   # 3.14
print(float(5))        # 5.0
print(float('inf'))    # inf       — special value!

# str() — converts to string
print(str(42))         # '42'
print(str(3.14))       # '3.14'
print(str(True))       # 'True'

# bool() — what evaluates to False?
print(bool(0))         # False     — zero is falsy
print(bool(''))        # False     — empty string is falsy
print(bool([]))        # False     — empty list is falsy
print(bool(None))      # False     — None is falsy
print(bool('0'))       # True      — non-empty string is TRUTHY!
```

```python
# input() ALWAYS returns str – even if user types a number
age = input('Enter age: ')  # user types 20
print(type(age))            # <class 'str'> – NOT int!
print(age + 1)              # TypeError: can only concatenate str to str

# Fix: convert immediately
age = int(input('Enter age: '))
print(age + 1) # 21 – now works
```

```python

# Safe pattern: combine with try/except
try:
    age = int(input('Enter age: '))
except ValueError:
    print('Please enter a whole number.')
print(age)
```

```python
# String + number – TypeError
#print('Score: ' + 95)        # TypeError!
print('Score: ' + str(95))   # 'Score: 95' – convert first
print(f'Score: {95}')        # 'Score: 95' – f-string handles it

# Comparing different types
print(1 == '1')     # False – int ≠ str even if same value
print(1 == True)    # True  – bool IS a subclass of int
print(0 == False)   # True  – False equals 0

# int vs float in operations
print(5 / 2)        # 2.5   – true division always gives float
print(5 // 2)       # 2     – floor division gives int
print(5.0 // 2)     # 2.0   – float // int gives float

# String repetition vs addition
print('3' + '4')    # '34'   – string concatenation
print('3' * 4)      # '3333' – string repetition
print(3 + 4)        # 7      – integer addition
print(3 * 4)        # 12     – integer multiplication
```

```python
# Trap 1: -2 ** 2
print(-2 ** 2)       # -4  (not 4!)  ** before unary -

# Trap 2: integer division sign
print(-7 // 2)       # -4  (rounds DOWN, not toward zero)

# Trap 3: modulo with negative
print(-7 % 3)        # 2   (sign follows divisor in Python)

# Trap 4: bool arithmetic
print(True + True)   # 2   (True == 1 in arithmetic)
print(True * 5)      # 5

# Trap 5: integer vs float division
print(type(4 / 2))   # <class 'float'>  (always float!)
print(type(4 // 2))  # <class 'int'>

# Trap 6: string * int
print('ha' * 3)      # 'hahaha' – repetition, not error

# Trap 7: int('3.5') is ValueError – must use float() first
print(int(float('3.5')))  # 3 – correct approach
```
