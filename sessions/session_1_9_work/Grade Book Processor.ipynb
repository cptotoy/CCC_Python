```python
grades = {'Alice':95,'Bob':72,'Carol':88,'Dave':55,'Eve':91}
# Print all student names (keys only)
print(grades.keys())
# Print all scores (values only)
print(grades.values())

#Use .items() to print each student with their score and letter grade (A≥90, B≥80, C≥70, D≥60, F<60)
for student, score in grades.items():
    print(f'{student}: {score}')

for student, score in grades.items():
    if score >= 90:
        print(f'{student} earned an A')
    elif score >= 80:
        print(f'{student} earned a B')
    elif score >= 70:
        print(f'{student} earned a C')
    elif score >= 60:
        print(f'{student} earned a F')

#Find and print the highest scorer and their score

highest_scorer = max(grades, key=grades.get)
highest_score = grades[highest_scorer]

print(f"Highest Scorer: {highest_scorer} with a score of {highest_score}")


# Create a new dict of only students who scored 60 or higher
passing_students = {student: score for student, score in grades.items() if score >= 60}

```
