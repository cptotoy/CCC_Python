#Creating Dictionaries - Syntax

```python
# Method 1: Curly braces (most common)
student = {
    'name': 'Alice',
    'age': 20,
    'grade': 'A',
    'gpa': 3.9
}

# Method 2: dict() constructor
config = dict(host='localhost', port=8080, debug=True)

# Empty dictionary - add items later
scores = {}
scores['math']     = 95  # adds key 'math' with value 95
scores['english'] = 87

# Keys must be UNIQUE - duplicate key overwrites the first
d = {'x': 1, 'x': 2}  # d = {'x': 2} - 1 is gone!
```

```python
student = {'name': 'Alice', 'age': 20, 'grade': 'A', 'gpa':3.5}

# Method 1: Square bracket - KeyError if key missing
print(student['name'])      # Alice
print(student['age'])       # 20
print(student['gpa'])     # KeyError: 'gpa' - crashes!

# Method 2: .get() - returns None (or default) if missing
print(student.get('grade'))        # A
print(student.get('gpa'))          # None - no crash
print(student.get('gpa', 0.0))     # 0.0  - custom default

# Check if key exists before accessing
if 'name' in student:
    print(student['name'])         # safe access

# PCEP trap: student['missing'] raises KeyError
# student.get('missing') returns None - no error
```

Adding, Updating & Deleting

```python
student = {'name': 'Alice', 'age': 20}

# ADD a new key-value pair
student['grade'] = 'A'        # {'name':'Alice','age':20,'grade':'A'}
print(student)

## UPDATE an existing value
student['age'] = 21           # overwrites 20 with 21
print(student)

## DELETE a key - del raises KeyError if missing
del student['grade']          # removes key entirely
print(student)

## .pop() - removes AND returns the value
age = student.pop('age')      # age = 21, key removed from dict
missing = student.pop('gpa', 0) # safe pop - returns 0 if missing
print(student)

## .update() - merge another dict in
extras = {'gpa': 3.9, 'year': 2}
student.update(extras)        # adds/overwrites with extras
print(student)

#Clear all items
student.clear()               # {} - empty dict, not deleted
print(student)
```
