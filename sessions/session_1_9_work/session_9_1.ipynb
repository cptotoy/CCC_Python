Essential Dictionary methods

```python
d = {'a': 1, 'b': 2, 'c': 3}

# .keys() - returns all keys as a view
print(d.keys())                 # dict_keys(['a', 'b', 'c'])

# .values() - returns all values as a view
print(d.values())               # dict_values([1, 2, 3])

# .items() - returns key-value pairs as tuples
print(d.items())                # dict_items([('a', 1), ('b', 2), ('c', 3)])

# Check membership - 'in' checks KEYS by default
print('a' in d)                 # True
print(1 in d)                   # False - 1 is a VALUE not a key
print(1 in d.values())          # True

# len() - number of key-value pairs
print(len(d))                   # 3
```

Iterating Over Dictionaries

```python
grades = {'Alice': 95, 'Bob': 87, 'Carol': 92}

# Iterate keys (default - same as 'for k in grades.keys()')
for student in grades:
    print(student)              # Alice Bob Carol

# Iterate values only
for score in grades.values():
    print(score)                # 95 87 92

# Iterate BOTH key and value - .items() is the PCEP favourite
for student, score in grades.items():
    print(f'{student}: {score}')
# Alice: 95
# Bob: 87
# Carol: 92

# Conditional inside dict iteration
for student, score in grades.items():
    if score >= 90:
        print(f'{student} earned an A')
```

Dictionary Comprenhensions

```python
# Syntax: {key_expr: value_expr for item in iterable}

# Build a dict of squares
squares = {n: n**2 for n in range(1, 6)}
# {1:1, 2:4, 3:9, 4:16, 5:25}

# Build from a list
students = ['Alice', 'Bob', 'Carol']
scores   = [95, 87, 92]
grade_book = {s: sc for s, sc in zip(students, scores)}
# {'Alice': 95, 'Bob': 87, 'Carol': 92}

# Filter: only passing grades (>= 60)
grades = {'Alice':95, 'Bob':55, 'Carol':72, 'Dave':45}
passing = {k: v for k, v in grades.items() if v >= 60}
# {'Alice': 95, 'Carol': 72}
print(passing)

# Transform values - convert Celsius to Fahrenheit
celsius = {'London': 15, 'Paris': 22, 'Tokyo': 28}
fahrenheit = {city: c*9/5+32 for city, c in celsius.items()}
```

```python
# Pattern 1: Frequency counter
words = ['cat', 'dog', 'cat', 'bird', 'dog', 'cat','cat']
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1
print(freq)
# {'cat': 3, 'dog': 2, 'bird': 1}

# Pattern 2: Group by category
students = [('Alice', 'A'), ('Bob', 'B'), ('Carol', 'A'), ('Dave', 'B')]
by_grade = {}
for name, grade in students:
    if grade not in by_grade:
        by_grade[grade] = []
    by_grade[grade].append(name)
# {'A': ['Alice', 'Carol'], 'B': ['Bob', 'Dave']}

# Pattern 3: Invert a dict (swap keys and values)
original = {'a': 1, 'b': 2, 'c': 3}
inverted = {v: k for k, v in original.items()}
# {1: 'a', 2: 'b', 3: 'c'}
```
